```python
import requests
from PIL import Image
```

```python
url = 'http://79.137.198.248:3000/classify'
img_path = './images/5379.jpg'

Image.open(img_path).resize((299,299))
```

```python
with open(img_path, 'rb') as f:
    data = f.read()
requests.post(url, data=data,headers= {"Content-Type": "image/xpm"}).json()
```
